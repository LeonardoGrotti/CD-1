## Introdução

Minha escolha para a análise desse projeto são os aposentados separados por genêro com os dados das PNADs de 2014 e 2011. O que me motivou na escolha dos grupo a serem analisados foi a indiscriminação sofrida pelas mulheres. Grande parte dos aposentados brasileiros nasceram na primeira metade do Século XX, um período em que o machismo era, por incrível que pareça, mais esbanjado do que hoje em dia. As variáveis escolhidas foram o sexo, idade, se a pessoa é aposentada ou não, renda mensal medida em salários mínimos e a renda recebida pela aposentadoria, nível de instrução mais elevado e se o entrevistado trabalhou na semana de referência.  

A pesquisa que utilizei foi uma do IPEA que compara os aposentados entre gêneros em alguns aspectos, utilizando variáveis quantitativas e qualitativas. (http://www.ipea.gov.br/agencia/images/stories/PDFs/politicas_sociais/05_capt02_7e.pdf)
Nessa pesquisa, algumas das analises feitas são os beneficiários do sstema de previdência por sexo e faixas etárias (gráfico 4), envolvendo aposentados e pensionistas; e uma analise dos números de aposentados por idade concedidos por sexo (gráfico 9)

```python
def scatter_and_line(dataframe, column1, column2):
    """ Recebe um dataframe e o nome de suas duas colunas, e faz
        o scatterplot dos dois e tenta achar a reta de tendência
    """
    a = dataframe[column1].cov(dataframe[column2]) / dataframe[column1].var()
    b = dataframe[column2].mean() - a * dataframe[column1].mean()

    dataframe.plot.scatter(x=column1, y=column2, color='red')

    import matplotlib.pyplot as plt
    plt.plot((dataframe[column1].min(), dataframe[column1].max()), (a * dataframe[column1].min() + b, a * dataframe[column1].max() + b), color='blue')
    plt.xlabel("Idade do Aposentado")
    plt.ylabel("Renda recebida pela aposentadoria")
    
    print("Coeficiente angular a: ", a)
    print("Coeficiente linear b: ", b)
```

```python
scatter_and_line(dadosnovos2014, "V8005", "V1252")
```

```python
scatter_and_line(dadosnovos2011, "V8005", "V1252")
```

Com os gráficos acima, foi possível analisar a dispersão entre a idade do aposentado e sua renda da aposentadoria recebida. Infelizmente, pode-se observar que quanto mais velho o aposentado, menos ele tende a receber de aposentadoria, uma vez que os coeficente angulares das linhas de tendência são negativos.

# Conclusão 

Depois de todas as analises feitas, algumas conclusões surgiram, dentre elas, as que mais se destacam, no que se trata de comparação entre os aposentados de cada sexo, são que os homens trabalham mais que as mulheres quando se aposentam, esse dado também aumentou de 2011 para 2014. Além disso, a diferença esperada entre as rendas dos homens e mulheres aposentadas não foi observada de acordo com as PNADs pesquisadas; vindas de uma época muito machista, as mulheres aposentadas possuem renda total muito semelhante à dos homens aposentados, tendo uma taxa de mulheres que completaram o ensino superior maior que a dos homens aposentados! As faixas de idade se comportaram de maneira semelhante entre os sexos, acompanhando a taxa de aumento de aposentados do ano de 2011 para 2014. Porém, o que pode ser concluido com relação a distribuição de renda de aposentadoria é que, quanto mais velho o aposentado, menos ele tende a receber de aposentadoria; esse tendência ficou ainda mais negativa no ano de 2014 do que em 2011. O que se pode inferir sobre isso é que o orçamento para aposentadorias de 2014 não acompanhou o índice de crescimento dos aposentados.
Portanto, dadas as circunstâncias, pode se dizer que as mulheres aposentadas brasileiras estão melhores em relação aos homens aposentados brasileiros, pois a taxa de mulheres aposentadas que trabalham é menor do que a de homens aposentados, suas rendas são semelhantes e a taxa de mulheres aposentadas que concluiram o ensino superior é maior do que a de homens aposentados.
