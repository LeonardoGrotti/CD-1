## 8)
### Deixo aqui minha tentativa final de plottar um QQplot
### Infelizmente, não soube fazer essse gráfico, haviam poucas informações sobre os modelos de distribuição que escolhi, uma vez que minha variável é discreta, e mesmo usando o código de exemplo fornecido pelos professores, não foi possível readaptar para o meu modelo

```python
# dados ordenados e quantis teóricos - QQplot

tempos = np.linspace(0, 43, 44)
# Calcula a frequência relativa acumulada com os dados vindos dos arquivos
fra = [] # frequencia relativa acumulada
for i in range(len(tempos)):
    acc = (i - 0.5) / len(tempos)
    if acc < 0:
        acc = 0
    fra.append(acc) # ver Magalhães e Lima (pag. 195)

# Convertemos fra em uma Series do Pandas
fra = pd.Series(data=fra)

# Gera os valores dos quantis teóricos para as distribuições analisadas
q_poisson = stats.poisson.pmf(tempos, fra ,loc=0)
q_nbinom = stats.nbinom.pmf(tempos, 0.254494, fra)

fig = plt.figure(figsize=(12,8))

plt.title("Q-Q plot: Dados x Poisson x Distrib. Binomial negativa")

poisson_plot = plt.plot(tempos, q_poisson, 'o', color="red", label="Exponencial")
nbinom_plot = plt.plot(tempos, q_nbinom, 'o', color="green", label = "Gamma")

#plt.legend(handles=[poisson_plot, nbinom_plot])
plt.plot((min(tempos), max(tempos)), (min(tempos), max(tempos)), '-')


for i in range(len(tempos)):
   plt.plot((tempos[i], tempos[i]), (q_poisson[i], tempos[i]), '-', color='red', alpha=0.3)
   plt.plot((tempos[i], tempos[i]), (q_nbinom[i], tempos[i]), '-', color='green', alpha=0.3)
```

## 9)
### Também não foi possível plottar esse próximo gráfico a partir do exemplo fornecido pelos professores, consegui essas duas curvas, mas que não se aproximam o suficiente da F.R.A

```python
cdf_exponencial = stats.poisson.cdf(tempos, 18) #FDA
cdf_gamma = stats.nbinom.cdf(tempos, 18, fra, loc=0)


fig = plt.figure(figsize=(12,8))

plt.title("Cumulativa: Dados x Exponencial x Gamma")

dados_fra_plot, = plt.plot(tempos, fra, '-', color='black', label="F.R.A dos dados")
cdf_exp_plot,   = plt.plot(tempos, cdf_exponencial, '-', color='red', label = "Poisson")
cdf_gamma_plot, = plt.plot(tempos, cdf_gamma, '-', color='green', label = "Distrib. Binom, Negativa")

plt.legend(handles=[dados_fra_plot, cdf_exp_plot, cdf_gamma_plot])


for i in range(len(tempos)):
    plt.plot((tempos[i], tempos[i]), (fra[i], cdf_exponencial[i]), ':', color="red", alpha=0.5)
    plt.plot((tempos[i], tempos[i]), (fra[i], cdf_gamma[i]), '--', color="green", alpha=0.5)
    
```
